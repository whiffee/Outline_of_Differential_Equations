<font color='blue'>12.1 Solve $\,y\,'''\,+\,y\,'\,=\,\textrm{sec}\,x\,$
</font>
</div>

<div style="background: #E2FFCC;

<font color='blue'>The entry is made into Wolfram Alpha:<br><br>
!| y''' + y' = sec(x) |!<br><br>
and the answer is received:<br><br>
$$y(x)=c_1\,\textrm{sin}\,(x)\,+\,c_2\,\textrm{cos}\,(x)\,+\,c_3-x\,\textrm{cos}\,(x)$$


$$-\textrm{log}\,\left(\textrm{cos}\,\left(\frac{x}{2}\right)\,-\,\textrm{sin}\,\left(\frac{x}{2}\,\right)\right)+\,\textrm{log}\,\left(\textrm{cos}\,\left(\frac{x}{2}\right)\,+\,\textrm{sin}\,\left(\frac{x}{2}\,\right)\right)+\,\textrm{sin}\,(x)\,\text{log}\,(\textrm{cos}\,(x))\,$$<br><br>

However, this answer does not match the text answer exactly. The text answer is:<br><br>

$$y\,=\,c_1\,+\,c_2\,\textrm{cos}\,x\,+\,c_3\,\textrm{sin}\,x\,+\,\textrm{log}\,|\,\textrm{sec}\,x\,+\,\textrm{tan}\,x\,|\,-\,x\,\textrm{cos}\,x\,+\,(\textrm{sin}\,x)\,\textrm{log}\,|\,\textrm{cos}\,x\,|\,$$<br>

Comparing the two answers, the significant difference seems to be in the question of whether:<br><br>

$$-\textrm{log}\,\left(\textrm{cos}\,\left(\frac{x}{2}\right)\,-\,\textrm{sin}\,\left(\frac{x}{2}\,\right)\right)+\,\textrm{log}\,\left(\textrm{cos}\,\left(\frac{x}{2}\right)\,+\,\textrm{sin}\,\left(\frac{x}{2}\,\right)\right)$$<br>

$$\stackrel{\large?}{\small=}$$<br>
$$\textrm{log}\,|\,\textrm{sec}\,x\,+\,\textrm{tan}\,x\,|\,$$<br>

or as simplified slightly, whether<br><br>

$$-\textrm{log}\,\left(\,|\,\textrm{cos}^2\,\left(\frac{x}{2}\right)\,-\,\textrm{sin}^2\,\left(\frac{x}{2}\,\right)\,|\,\right)$$<br>

$$\stackrel{\large?}{\small=}$$<br>
$$\textrm{log}\,|\,\textrm{sec}\,x\,+\,\textrm{tan}\,x\,|\,$$<br>

At present, the answer seems to be in the negative. A plot is added to show what seems to be the distinction between the functions. In the plot, all constants were assigned a value of 1.


                                                                                                     
                                                                                                     



</font>                 
</div>

```python

import numpy as np
import matplotlib.pyplot as plt

%config InlineBackend.figure_formats = ['svg']

x = np.linspace(0,10,300)
y1 = -np.log(abs((np.cos(x/2))**2 - (np.sin(x/2))**2))
y2 = np.log(abs(1/np.cos(x) + np.tan(x)))

        
plt.grid(True, linestyle='dotted', color='gray', linewidth=0.9)
plt.xlabel("x")
plt.ylabel("y(x)")
plt.title("Outline of Differential Equations Prob 12.1")
plt.rcParams['figure.figsize'] = [9, 7.5]
        
ax = plt.gca()
#ax.axhline(y=0, color='#993399', linewidth=1)
ax.axvline(x=0, color='#993399', linewidth=1)

#plt.text(965, -2.5e43, "-np.log(abs((np.cos(x/2))**2 - (np.sin(x/2))**2)))
#plt.text(965, 2.25e43, "SOLN: y = 3*np.exp(x**2) + 1/2", size=10,\
        # bbox=dict(boxstyle="square", ec=('#8C564B'),fc=(1., 1., 1),))
plt.ylim(-8,8.5)
plt.plot(x, y1, linewidth = 0.9)
plt.plot(x, y2, linewidth = 0.9)
plt.show()

```

<font color='blue'>12.2 Solve $\,y\,'''\,-\,3\,y\,''\,+\,2y\,'\,=\,\large\frac{e^{x}}{1\,+\,e^{-x}}\small\,$
</font>
</div>

<div style="background: #E2FFCC;
            font-family: Bookerly;
            font-size: 16px; 
            font-weight: 100;
            padding: 10px 5px 10px 5px; 
            border: 10px solid #E2FFCC;

<font color='blue'>The entry is made into Wolfram Alpha:<br><br>
!| y''' - 3* y'' + 2 * y' = e^x/(1 + e^(-x)) |!<br><br>
and the answer is received:<br><br>
$$\,y(x)\,=\,c_1\,\frac{e^{x}}{2}\,+\,\frac{1}{2}c_2\,e^{2x}\,+\,c_3\,+\,\frac{1}{2}\,e^{2x}\,x\,\hspace{24 em}$$ 


$$-\,e^x\,\textrm{log}\,(e^x\,+1)\,-\,\frac{1}{2}\,e^{2x}\,\,\textrm{log}\,(e^x\,+\,1)\,-\,\frac{1}{2}\,\textrm{log}\,(e^x\,+\,1)\hspace{20 em}$$<br>


The answer shown above does not match that of the text. The answer in the text goes like this:<br><br>
$$y(x)\,=\,c_1\,+\,c_2\,e^x\,+\,c_3\,e^{2x}\,+\,\frac{1}{2}\,(e^x\,+\,1)\,-\,\frac{1}{2}\,\textrm{log}\,(e^x\,+\,1)\,\hspace{18 em}$$
$$-\,e^x\,\textrm{log}\,(e^x\,+\,1)\,-\,\frac{1}{2}\,e^{2x}\,\textrm{log}\,(1\,+\,e^{-x})\,\hspace{26 em}$$<br><br>
Both versions of the answer have seven terms. But terms 4 and 6 in the top version do not have analogous terms, with the same basis, in the bottom version. One possible check (though not conclusive), is to compare plots of the functions. All constants were assigned a value of 1 in the test. Evidently there is a real disagreement about this solution.

Read on for the surprising dénouement.




</font>                 
</div>

```python
import numpy as np
import matplotlib.pyplot as plt

%config InlineBackend.figure_formats = ['svg']

x = np.linspace(-10,10,300)
y3 = np.exp(x)/2 + (1/2)*np.exp(2*x) +1 + (1/2)*np.exp(2*x)*x \
    -np.exp(x)*np.log(np.exp(x)+1) -(1/2)*np.exp(2*x)*np.log(np.exp(x) \
    + 1) -(1/2)*np.log(np.exp(x) + 1)
y4 = 1 + np.exp(x) + np.exp(2*x) + (1/2)*(np.exp(x) \
    + 1) - (1/2)*(np.log(np.exp(x) + 1)) \
    - np.exp(x)*np.log(np.exp(x) + 1) - \
    (1/2)*np.exp(2 * x)*np.log(1 + np.exp(-x))
@np.vectorize
def flatWA(x):
    return 0.61370563888

@np.vectorize
def flatTx(x):
    return 2.61370563888

plt.grid(True, linestyle='dotted', color='gray', linewidth=0.9)
plt.xlabel("x")
plt.ylabel("y(x)")
plt.title("Outline of Differential Equations Prob 12.2")
plt.rcParams['figure.figsize'] = [9, 7.5]

        
ax = plt.gca()
ax.axhline(y=0, color='#993399', linewidth=1)
ax.axvline(x=0, color='#993399', linewidth=1)
ratio = 1.0
xleft, xright = ax.get_xlim()
ybottom, ytop = ax.get_ylim()
ax.set_aspect(abs((xright-xleft)/(ybottom-ytop))*ratio)


plt.text(0, 0, "Wolfram Alpha soln", size=10,\
         bbox=dict(boxstyle="square", ec=('#1F77B4'),fc=(1., 1., 1),))
plt.text(-2.5, 2.5, "Text's soln", size=10,\
         bbox=dict(boxstyle="square", ec=('#FF1D1D'),fc=(1., 1., 1),))
plt.ylim(-2,8)
plt.xlim(-6,4)
plt.plot(x, y3, linewidth = 0.9)
plt.plot(x, y4, linewidth = 0.9, color = 'r')
plt.plot(x, flatWA(x), linewidth = 0.9, color = 'g', linestyle = 'dashed')
plt.plot(x, flatTx(x), linewidth = 0.9, color = 'y', linestyle = 'dashed')
plt.show()

```
