<font color='blue'>7.1 A ${$}$20,000 savings account pays 5 percent interest per year, compounded continuously. Find $\,(a)\,$ the amount in the account after three years, and $\,(b)\,$ the time required for the account to double (if static).

<font color='blue'>$\,N\,$ represents the current balance. At first, $\,N(0)\,=\,20,000\,$. The balance grows by interest payments, in proportion to the amount in the account. The constant of proportionality is $\,k\,=\,0.05\,$ and the relationship is $\,\large\frac{dN}{dt}\,\small-\,0.05N\,=\,0\,$. Entering this expression in Wolfram Alpha yields <br><br>
$$\,N(t)\,=\,c_1e^{0.05t}\,$$<br>
At time $\,t\,=\,0,\,\:\:N(0)\,=\,20000\,$, and <br><br>
$$\Longrightarrow 20000\,=\,c_1\,$$<br>
Substituting upward produces <br><br>
$$\Longrightarrow \,N(t)\,=\,20000e^{0.05t}$$<br>
and the last is a description of the balance at any time $\,t\,\hspace{1 em}( \textrm{with}\,t\,\textrm{in years})\,$<br><br>
Plugging in 3 for $\,t\,$, and putting the rhs into Wolfram Alpha gives:<br><br>
$$\,{$}23,236.68\,$$
which answers $\,(a)\,$.<br><br>
When will the account equal 40000?<br><br>
$$\,40000\,=\,20000e^{0.05t}$$<br>
$$\Longrightarrow 2\,=\,e^{0.05t}$$<br>
$$\Longrightarrow \textrm{log (abs(2))}\,=\,0.05t\,$$<br>
In Wolfram Alpha, the entry line is:<br><br>
!| log(abs(2)) = 0.05t, t |!<br><br>
and the answer is 13.86, which answers $\,(b)\,$

<font color='blue'>7.2 A bank account of $\,{$}5000\,$ earns interest compounded continuously. If undisturbed, what will the account be worth after 7 years if the interest rate is 8.5 percent for the first 4 years and 9.25 percent for the last three years?
</font>
</div>

<div style="background: #E2FFCC;
            font-family: Bookerly;
            font-size: 16px; 
            font-weight: 100;
            padding: 10px 5px 10px 5px; 
            border: 10px solid #E2FFCC; 
            margin-left: 1px;
            margin-right: 1px;"

<font color='blue'>Having two different interest rates in the same problem effectually turns it into two problems.<br><br>
Let $\,N(t)\,$ be the balance in the account at any time $\,t\,$. At the beginning, $\,N(0)\,=\,5000\,$. For the first 4 years, $\,k\,=\,0.085\,$ and <br><br>
$$\,\frac{dN}{dt}\,-\,0.085\,N\,=\,0\,$$<br>
Entering this expression into Wolfram Alpha<br><br>
!| dN/dt - 0.085 N = 0 |! <br><br>
produces the following
$$\,N(t)\,=\,c_1\,e^{0.085t}\,$$<br>
At time $\,t\,=\,0\,,\:\:N(0)\,=\,5000\,$, and both of these initial conditions can be substituted into the previous equation, changing it to<br><br>
$$\,5000\,=\,c_1e^{0.085\,(0)}\,=\,c_1\,\hspace{2 em}(0\,\le t\,\le 4)$$<br>
and 5000 can take the place of $\,c_1\,$ in the previous equation. Wanting to know the value of the bank account after 4 years means entering the line:<br><br>
!| N(t) = 5000 e^(0.085 4) |!<br><br>
into Wolfram Alpha. The response is an answer of 7024.74. So that is the signpost at the 4-year mark. The second part of the problem can be started now.<br><br> The interest rate increases to 9.25 percent, which means entering the following line in Wolfram Alpha:<br><br>
!| dN/dt - 0.0925 N = 0 |!<br><br>
The answer given, in a line that can be referred to as "4plus" is:<br><br>
$$\,N(t)\,=\,c_1\,e^{0.0925\,t}\,$$<br>
Now the 4plus line can be modified to read:<br><br>
$$\,7024.74\,=\,c_1\,e^{0.0925\,(4)}$$<br>
Consulting W|A,<br><br>
!| 7024.74 = c1 e^(0.0925 4), c1 |!<br><br>
it is found that $\,c_1\,=\,4852.23\,$, and therefore the calculation equation has become<br><br>
$$\,N(t)\,=\,4852.23\,e^{0.0925\,t}\,$$<br>
Only one more calculation is required, and the entry in W|A is made:<br><br>
!| 4852.23 e^(0.0925 * 7) |!<br><br>
resulting in the answer, in dollars, for the whole 7-year period: 9271.44. Using the full period here instead of only the last 3 years is not intuitive, and suggests the following test:<br><br>
!| 4852.23 * e^(0.0925 * 4) |!<br><br>
which results in the answer: 7024.44, exactly the same amount which the original $\,{$5000}\,$ became under the original lower interest rate after 4 years. So in a sense the value 4852.23 seems a fictitious one, invented so that the 2-step process can be feigned to occur in only 1 step.

<font color='blue'>The account balance is described by<br><br>
$$\,\frac{dN}{dt}\,-\,kN\,=\,0\,$$<br>
And entering this expression into Wolfram Alpha results in the answer:<br><br>
$$\,N(t)\,=\,c_1\,e^{kt}\,$$<br>
At time $\,t\,=\,0\,$ the starting balance is $\,N_0,\,\:$ though what that amount is remains unknown. What is known is that<br><br>
$$\,N_0\,=\,c_1\,e^{k\,(0)}\,=\,c_1\,$$<br>
and $\,N_0\,=\,c_1\,$ means that<br><br>
$$\,N(t)\,=\,N_0\,e^{kt}\,$$<br>
What the problem wants is the rate of interest, k, that will allow $\,N_0\,$ to double in six years. Ideally, Wolfram Alpha could do this in one step with the line:<br><br>
!| N_0 = 2 * N_0 * e^(6 * k * t), k |!<br><br>
Unfortuately, W|A has no clue to what this multi-variable snipe hunt could mean. It is necessary to pry things apart by hand first. For the sake of unconfusing nomenclature for W|A, proceed in this way:<br><br>
$$\,2N_0\,=\,N_0e^{6k}\,$$<br>
$$\,\Longrightarrow e^{6k}\,=\,2\,$$<br>

Then apply a log identity to get:<br><br>
$$\,{6*k = log |2|}\,$$<br>
Then enter into W|A the following:<br><br>
!| N[6*k = log|2|], k |!<br><br>
To encourage W|A to respond with numerical output. The answer is k = 0.115525; that is, an interest rate of 11.5 percent would be required to get the balance doubled in six years.<br><br>

Plugging these numbers in produces the equation<br><br>

<font color='blue'>7.7 Five mice in a stable population of 500 are intentionally infected with a contagious disease to test a theory of epidemic spread that postulates the rate of change in the infected population is proportional to the product of the number of mice who have the disease with the number that are disease free. Assuming the theory is correct, how long will it take half the population to contract the disease?
</font>
</div>

<div style="background: #E2FFCC;
            font-family: Bookerly;
            font-size: 16px; 
            font-weight: 100;
            padding: 10px 5px 10px 5px; 
            border: 10px solid #E2FFCC; 
            margin-left: 1px;
            margin-right: 1px;"

<font color='blue'> N(t) will be the number of mice with the disease at time $\,t.\,$ At $\,N(0)\,$ the number of infected is 5, and the number of uninfected will be $\,500\,-\,N(t).\,$ The theory predicts that<br><br>
$$\,\frac{dN}{dt}\,=\,k\,N(500 - N)\,$$<br>
The character $\,k\,$ represents the constant of proportionality. The proportionality here relates to the number of disease free mice. The last equation can be rearranged:<br><br>
$$\,\frac{dN}{N\,(500\,-\,N)}\,-\,k\,dt\,=\,0\,$$<br>
Wolfram Alpha can process this equation using partial fraction decomposition:<br><br>
!| [ 1/(N*(500 - N)) ] /Apart |!<br><br>
which results in the expression:<br><br>
$$\,\frac{1}{500 N}\,-\,\frac{1}{500\,(N\,-\,500)}\,$$<br>
which can be rearranged in the somewhat unorthodox way:<br><br>
$$\,\frac{\frac{1}{500}}{N}\,+\,\frac{\frac{1}{500}}{ 500\,-\,N}\,$$<br><br>
and then joined up with the part concerning the $\,k\,$ factor to give:<br><br>
$$\,\frac{1}{500}\,\left(\frac{1}{N}\,+\,\frac{1}{500\,-\,N}\right)\,dN\,-\,k\,dt\,=\,0\,$$<br>
To get the solution to this equation, in Wolfram Alpha the line can be entered:<br><br>
!| 1/500 * integrate[1/N + 1/(500 - N) dN] - integrate[k dt] |!<br><br>
which results in:<br><br>
$$\,\frac{1}{500}\,(\textrm{log}\,|\,N\,|\,-\,\textrm{log}\,|\,500\,-\,N\,|\,)\,-\,kt\,=\,c\,$$<br>
(Constant c and absolute value bars added manually to the last equation.)<br><br>
$$\Longrightarrow \textrm{log}\left|\frac{N}{500\,-\,N}\right|\,=\,500(c\,+\,kt)
,$$<br>
$$\,\Longrightarrow \frac{N}{500\,-\,N}\,=\,e^{500\,(c\,+\,kt)}\,$$<br>
$$\,\Longrightarrow \frac{N}{500\,-\,N}\,=\,c_1\,e^{500\,kt}\,$$<br>
At time $\,t\,=\,0, \,N\,=\,5,\,$ and substituting this condition shows:<br><br>
$$\,\frac{5}{495}\,=\,c_1\,e^{500\,k\,(0)}\,=\,c_1\,$$<br><br>
So the general equation for calculating the number of infected is modified by the substitution of the above to become:<br><br>
$$\,\frac{N}{500 \,-\,N}\,=\,\frac{1}{99}\,e^{500\,kt}\,$$<br>
The situation we are tasked to find out about is the one with 250 infected mice. This adjusts the general equation to point to:<br><br>
$$\,1\,=\,\frac{1}{99}\,e^{500\,kt}\,$$<br>
This equation can be solved directly by Wolfram Alpha, by entering the line:<br><br>
!| N[1 = 1/99 * e^(500 * k * t)], t |!<br><br>
and the answer delivered is:<br><br>
$$\,t\,\approx\,\frac{0.00919024}{k}\,$$<br>

!| dQ/dt + (2/(10 + 2 * t)) * Q = 4 |!<br><br>

<font color='blue'>The basic equation governing the amount of current $\,I\,$ (in amperes) in a simple RL circuit consisting of a resistance $\,R\,$ (in ohms), an inductor $\,L\,$ (in henries), and an electromotive force (abbreviated emf) $\,E\,$ (in volts) is:<br><br>
$$\frac{dI}{dt}\,+\,\frac{R}{L}\,I\,=\,\frac{E}{L}\,$$<br>
In this particular problem, $\,E\,=\,5, R\,=\,50,\,$ and $\,L\,=\,1\,$, so the EMF steady state equation becomes:<br><br>
$$\,\frac{dI}{dt}\,+\,50I\,=\,5\,$$<br><br>
The ODE is fed to Wolfram Alpha with the line<br><br>
!| dC/dt +50 * C = 5 |!<br><br>
and the returned answer is:<br><br>
$$\,I(t)\,=\,c_1\,e^{-50t}\,+\,\frac{1}{10}\,$$<br><br>
Wolfram Alpha reveals a wart in this procedure, because introduction of the $\,'I'\,$ character, even capitalized, is taken for Euler's imaginary $\,i\,$. So an alternate character, as above, needs to be substituted, then retracted on exit.<br><br>
Initial conditions will solidify the general equation for current. At time $\,t\,=\,0\,\,I\,=\,0\,.$ Applying this condition to the above equation reveals that the value of $\,c_1\,$ must be zero. Therefore, the current at any time is equal to:<br><br>
$$\,I\,=\,-\frac{I}{10}\,e^{-50t}\,+\,\frac{1}{10}\,$$<br>
The above equation is the answer to the question posed by the problem.

<div style="background: #E2FFCC;

<font color='blue'>Consider a one-parameter family of curves in the $\,xy-$plane defined by<br><br>
$$\,F(x, y, c)\,=\,0\,$$<br>
where $\,c\,$ stands for the parameter. The problem is to find another one-parameter family of curves, called the orthogonal trajectories of the family cited above, and given analytically by<br><br>
$$\,G(x, y, k)\,=\,0\,$$<br>
such that every curve in the new family intersects at right angles with every curve in the original family.<br><br><br>
The family, which is given by $\,F(x, y, c) \,=\,x^2\,+\,y^2\,-\,c^2\,$ is made up of circles whose centers are at the origin and whose radii are given by $\,c\,$. Wolfram Alpha will do an implicit differentiation with the entry:<br><br>
!| implicit differentiation [x^2 + y^2 = c^2] |!<br><br>
returning the expression<br><br>
$$\,y\,'(x)\,=\,-\frac{x}{y}\,$$<br>
Orthogonal trajectories are solutions of the equation<br><br>
$$\,\frac{dy}{dx}\,=\,-\frac{1}{f(x,y}\,$$<br>
It is the procedure with o.p. problems to process the above form into<br><br>
$$\,\frac{dy}{dx}\,=\,\frac{y}{x}\,$$<br>
The ODE just shown can be solved by Wolfram Alpha routinely:<br><br>
!| dy/dx = y/x |!<br><br>
The solution is:<br><br>
$$\,y(x)\,=\,c_1x\,$$<br>
Next comes a plot of some representative family members along with some orthogonal projections. Note that as required, intersections occur at right angles. The parametric plot style is convenient.



</font>                 
</div>

```python
import numpy as np 
import matplotlib.pyplot as plt 

%config InlineBackend.figure_formats = ['svg']

x = np.linspace(-10,10,1000)
y1 = -x
y2 = x
y3 = 2*x

angle = np.linspace( 0 , 2 * np.pi , 150 ) 
fig, axes = plt.subplots( 1 ) 
ax = plt.gca()
ax.axhline(y=0, color='#993399', linewidth=1)
ax.axvline(x=0, color='#993399', linewidth=1)

    
plt.text(1.8, -2, "y = -x", size=10,bbox=dict\
         (boxstyle="square", ec=('#D62728'),fc=(1., 1., 1),))
plt.text(1.8, 2, "y = x", size=10,bbox=dict\
         (boxstyle="square", ec=('#9467BD'),fc=(1., 1., 1),))
plt.text(1.1, 2.75, "y = 2x", size=10,bbox=dict\
         (boxstyle="square", ec=('#86564B'),fc=(1., 1., 1),))
plt.text(0.5, 0.15, "x^2 + y^2 = 1", size=8,bbox=dict\
         (boxstyle="square", ec=('#1F77B4'),fc=(1., 1., 1),))
plt.text(1.5, 0.75, "x^2 + y^2 = 2", size=8,bbox=dict\
         (boxstyle="square", ec=('#FF7F0E'),fc=(1., 1., 1),))
plt.text(2.65, 2.3, "x^2 + y^2 = 4", size=8,bbox=dict\
         (boxstyle="square", ec=('#2CA02C'),fc=(1., 1., 1),))
        
#radius = 1.0
plt.rcParams['figure.figsize'] = [8, 8]

axes.plot(1.0*np.cos( angle ), 1.0*np.sin( angle ), linewidth = 0.9)
axes.plot(2.0*np.cos( angle ), 2.0*np.sin( angle ), linewidth = 0.9)
axes.plot(4.0*np.cos( angle ), 4.0*np.sin( angle ), linewidth = 0.9)

axes.plot(x, y1, linewidth = 0.9, linestyle='dashdot')
axes.plot(x, y2, linewidth = 0.9, linestyle='dashdot')
axes.plot(x, y3, linewidth = 0.9, linestyle='dashdot')

axes.set_aspect( 1 )
axes.grid(True, linestyle='dotted', color='gray', linewidth=0.9)
plt.xlim(-4, 4)
plt.ylim(-4,4) 
plt.title( 'Outline of Differential Equations Prob 7.23' ) 
plt.show()
```

<font color='blue'>Consider a one-parameter family of curves in the $\,xy-$plane defined by<br><br>
$$\,F(x, y, c)\,=\,0\,$$<br>
where $\,c\,$ stands for the parameter. The problem is to find another one-parameter family of curves, called the orthogonal trajectories of the family cited above, and given analytically by<br><br>
$$\,G(x, y, k)\,=\,0\,$$<br>
such that every curve in the new family intersects at right angles with every curve in the original family.<br><br><br>
The family, which is given by $\,F(x, y, c) \,=\,y\,-\,cx^2\,$ is made up of parabolas. Wolfram Alpha will do an implicit differentiation with the entry:<br><br>
!| implicit differentiation [y = c * x^2] |!<br><br>
returning the expression<br>
$$\,y\,'(x)\,=\,2cx\,$$<br>
At this point it is necessary to get rid of the constant $\,c\,$.<br><br>
Note that this is not an arbitrary constant added to an integral, but an actual element of the family inventory. So the original equation can be solved to give $\,c\,=\,\large\frac{y}{x^2}\small\,$. Then the substitution of the rhs can be made into the derivative, yielding $\,\large\frac{dy}{dx}\small\,=\,\large\frac{2y}{x}\,\small$. This is where the procedure does that weird flipping+sign change thing, giving<br><br>
$$\frac{dy}{dx}\,=\,\frac{-x}{2y}\,$$<br>
or<br><br>
$$\,x\,dx\,+\,2y\,dy\,=\,0\,$$<br>


which is another ODE for Wolfram Alpha to solve.<br><br>
!| x * dx + 2 * y * dy = 0 |!<br><br>
and Wolfram Alpha comes back with:<br><br>
$$\,y(x)\,=\,-\frac{\sqrt{c_1\,-\,x^2}}{\sqrt{2}}\,\hspace{4 em}\textrm{and}\hspace{4 em}\,y(x)\,=\,\frac{\sqrt{c_1\,-\,x^2}}{\sqrt{2}}\,$$<br>
Testing out the solutions devised by Wolfram Alpha is shown in the multi-curve plot below. As far as native visual judgment goes, the intersections look square. Not all the ellipses would plot cleanly, making it necessary to reject some which were tested.




</font>                 
</div>
