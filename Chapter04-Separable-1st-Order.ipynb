```python
from sympy import *
x, y, c1 = symbols('x y c1', real=True)
expr1=x
xup2=integrate(expr1,x)
xup2
```

```python
eq1 = Eq(xup2 + yup3, c1)
eq1
```

```python
result = solve([eq1],(y))
result
```

```python
from sympy import *
x, y, c1 = symbols('x y c1')
expr1=-x**3
xup3=integrate(expr1,x)
xup3
```

```python
expr2= y**-2
yup2=integrate(expr2,y)
yup2
```

```python
eq1 = Eq(yup2 + xup3, c1)
eq1
```

<font color='blue'>Some changes are needed to get to the differential form. To solve this problem, some substitutions from homogeneous attributes need to be applied. That is,<br><br>
$$\,y\,=\,xv\,$$<br>
$$\,\frac{dy}{dx}\,=\,v\,+\,x\frac{dv}{dx}\,$$<br>
Applying these substitutions to the given problem. The two equations above will be referred to as the h.equations. The rhs of the lower h.equation is substituted for the $\,y\,'\,$ of the problem equation. Then the rhs of the upper h.equation is substituted for the rhs $\,y\,$ in the problem equation, bringing it to:<br><br>
$$\hspace{13 em}\,v\,+\,x\frac{dv}{dx}\,=\,\frac{xv\,+\,x}{x}\,\hspace{12 em}(1)$$<br>
The left hand $\,v\,$ of equation (1) can be brought to the rhs, giving (2).<br><br>
$$\hspace{13 em}\,x\frac{dv}{dx}\,=\,\frac{xv\,+\,x}{x}\,-\,v\,\hspace{12 em}(2)$$<br>
Multiplying this new $\,v\,$ by $\,\large\frac{x}{x}\normalsize\,$ and combining the rhs, converts the whole 
rhs to 1 as in (3).<br><br>
$$\hspace{15 em}\,x\frac{dv}{dx}\,=\,1\hspace{14 em}(3)$$<br><br>
Now moving the lhs x factor to rhs denominator, then multiplying remaining lhs denominator across, then subtracting remaining lhs factor, then mirroring, completely unpacks (3), giving (4).<br><br>
$$\hspace{14 em}\,\frac{1}{x}\,dx\,-\,dv\,=\,0\,\hspace{13 em}(4)$$<br>
The homogeneous-enabled substitutions have transformed the problem equation into one that is separable and ready to work on.


     
     
</font>                 
</div>
