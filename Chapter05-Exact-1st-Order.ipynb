<font color='blue'>The equation having passed the exactness test, it is now time to seek solutions to the two blobs that make up the equation. The outcome of such searching, should it be successful, will constitute the function of two variables, g(x,y). The target conditions of the two blobs consist in:<br><br>
$$\frac{\partial g(x,y)}{\partial x}\,=\,M(x,y)\,$$<br>
and<br>
$$\frac{\partial g(x,y)}{\partial y}\,=\,N(x,y)\,$$<br><br>
Using the name tag from Problem 5.1,$\,M(x,y)\,=\,2xy\,$, <br><br>
$$\,\frac{\partial g(x,y)}{\partial x}\,=\,2y\,$$<br>
and if the reverse operation is applied to $\,M\,$, (with respect to x) the following expression may be used<br><br>
$$\,\int\,\frac{\partial g(x,y)}{\partial x}\,=\,\int\,2xy\: dx\,$$<br>
$$\hspace{-3 em}\Longrightarrow\hspace{2 em}g(x,y)\,=\,x^2y\,+\,h(y)\,$$<br>
Since $\,g(x,y)\,$ is a function of two variables, differentiation with respect to $\,y\,$ is not the reverse of the process that was just performed, and <br><br>
$$\hspace{-3 em}\Longrightarrow\hspace{2 em}\frac{\partial g(x,y)}{\partial y}\,=\,x^2\,+\,h\,'(y)\,$$<br>
Choosing from available resources, the value of $\,h(y)\,$ can be found. The above expression is set adjacent to the $\,N(x,y)\,$ function, which it must equal.<br><br>
$$\hspace{-3 em}\Longrightarrow\hspace{2 em}\,x^2\,+\,h\,'(y)\,=\,1\,+\,x^2\,$$<br>
readily yielding $h\,'(y)\,=\,1\,$. Now the rhs quantity, above, can be integrated with respect to $\,y\,$: <br><br>
$$\,\int\,1\,+\,x^2\: dy\,=\,g(x,y)\,=\,x^2y\,+\,y\,+\,c_1$$<br><br>


     
     
     

</font>                 
</div>

<font color='blue'>In the chapter intro it was posited that the pure form of $\,g(x,y)\,$ would be a constant, $\,c\,$. This putative constant may now be borrowed to stand in the equation<br><br>
$$c\,=\,x^2y\,+\,y\,+\,c_1\,$$<br>
and if $\,c_2\,$ is chosen to equal $\,c\,-\,c_1\,$, then <br><br>
$$\,x^2y\,+\,y\,=\,c_2\,$$<br>
or if $\,y\,$ is sought exclusively,<br><br>
$$\,y\,=\,\frac{c_2}{x^2\,+\,1}\,$$<br>
Some example plots in the family of solutions just developed may be of interest.


     
     
     

</font>

```python
import numpy as np
import matplotlib.pyplot as plt

%config InlineBackend.figure_formats = ['svg']

x=np.linspace(-10,10,500)

two = np.array([-3,-2,-1, 1, 2, 3])

def f(x):
    for k in two:
        y = k/(x**2 + 1)
        plt.plot(y, linewidth = 0.9)
y=f(x)
        
plt.grid(True, linestyle='dotted', color='gray', linewidth=0.9)
plt.xlabel("x")
plt.ylabel("y(x)")
plt.title("Outline of Differential Equations Prob 5.2")
plt.rcParams['figure.figsize'] = [7, 7]
        
ax = plt.gca()
ratio = 1.2
#ratio is adjusted by eye to get squareness of x and y spacing
xleft, xright = ax.get_xlim()
ybottom, ytop = ax.get_ylim()
ax.set_aspect(abs((xright-xleft)/(ybottom-ytop))*ratio)
ax.axhline(y=0, color='#993399', linewidth=1)
ax.axvline(x=0, color='#993399', linewidth=1)
ax.set_xticks([0, 100, 200, 300, 400, 500], 
              labels=['0', '2', '4', '6', '8', '10' ])
ax.set_yticks([-3, -2, -1, 0, 1, 2, 3]) 
             # labels=['0', '1',  '2', '3',  4,  5,  6 ])

plt.text(29, -2.9, "SOLN: y = -3/(x**2 + 1)", size=10,bbox=dict\
         (boxstyle="square", ec=('#1F77B4'),fc=(1., 1., 1),))
plt.text(275, 2.6, "SOLN: y = 3/(x**2 + 1)", size=10,bbox=dict\
         (boxstyle="square", ec=('#8C564B'),fc=(1., 1., 1),))

```

<font color='blue'> The equation having been shown to be exact, the function of two variables $\,g(x,y)\, = \,c\,$, with derivatives equalling the M and N that have just been seen, will be sought. Trotting out the M introduced for this equation, and setting it up for integration, with respect to $\,x\,$<br><br>
$$\int\,\frac{\partial g}{\partial x}\,dx\,=\,\int\,(x\,+\,\textrm{sin}\,y)\,dx\,$$<br>
and following through with the integration<br><br>
$$\,g(x,y)\,=\,\frac{1}{2}\,x^2\,+\,x\,\textrm{sin}\,y\,+\,h(y)\,$$<br>
The above, here designated "the unfinished", is not the most specific description that can be had about g(x,y). What is needed at this point is the exact value of h(y). To that end, the last equation is differentiated with respect to $\,y\,$.
     
     
     

</font>                 
</div>

```python
from sympy import *
x, y, g, h = symbols('x y g h')
h = Function("h")(y)
hp = diff(h, y)
hp

```

```python
g = (1/2*x**2 + x*sin(y))
gp = diff(g, y)
gp

```

```python
gpa = gp + hp
gpa

```

The integration with respect to $\,y\,$ is a natural next move, bringing $\,h(y)\,$ into the form $\,h(y)\,=\,-\,y^2\,+\,c_1\,$.  At this point it makes sense to substitute $\,h(y)\,$ into the "unfinished" formulation above, and if this is done, then it is seen that<br><br>
$$\,g(x,y)\,=\,\frac{1}{2}\,x^2\,+\,x\,\textrm{sin}\,y\,-\,y^2\,+\,c_1\,$$<br>
Since the orignal assertion was that $\,g(x,y)\,=\,c\,$, then by assigning $\,c_2\,=\,c\,-\,c_1\,$, a final expression for the solution is given as<br><br>
$$\,\frac{1}{2}\,x^2\,+\,x\,\textrm{sin}\,y\,-\,y^2\,=\,c_2\,$$<br>

<font color='blue'> This equation needs to be tested for exactness. It therefore needs to be put into the form where M, N, and the indigenous differentials can be examined. This would look like:<br><br>
$$\,\hspace{1 em}(2y\,-\,x\,\textrm{exp}(xy))\,dy\,=\,(2\,+\,y\,\textrm{exp}(xy))\,dx\,$$<br>
Except that by convention both blobs are put on the same side of the equals sign, like<br><br>
$$\,\hspace{2 em}\Longrightarrow\,(2\,+\,y\,\textrm{exp}(xy))\,dx\, +\,(-\,2y\,+\,x\textrm{exp}(xy))\,dy\,=\,0\,$$<br>
The left hand blob is M, so its partial is taken with respect to $\,y\,$, and equals<br><br>
$$\,\textrm{exp}\,(xy)\,+\,xy\,\textrm{exp}\,(xy)\,$$<br>
The right hand blob is N, so its partial is taken with respect to $\,x\,$, and equals<br><br>
$$\,\textrm{exp}\,(xy)\,+\,xy\,\textrm{exp}\,(xy)\,$$<br>

</font>                 
</div>

<div style="background: #E2FFCC;
            font-family: Bookerly;

            padding: 20px 5px 20px 5px; 
            border: 30px solid #E2FFCC; 
            margin-left: 1px;
            margin-right: 1px;"

<font color='blue'>The equality of the above two lines means that the equation is exact.

Blob M, instead of being differentiated with respect to y, is now integrated with respect to x, yielding<br><br>
$$\,\int\,[2\, +\,y\, \textrm{exp}(xy)]\,dx\,$$<br>
$$\hspace{2 em}\Longrightarrow g(x,y)\,=\,2x\,+\,\textrm{exp}(xy)\,+\,h(y)\,$$<br>
which may be referred to as the "unfinished g". If this last line is differentiated with respect to $\,y\,$, the expression<br><br>
$$\,x\,\textrm{exp}(xy)\,+\,h\,'(y)\,$$<br>
is obtained. There is a chained obligation for this expression to equal blob N, thus<br><br>
$$\,x\,\textrm{exp}(xy)\,+\,h\,'(y)\,=\,x\,\textrm{exp}(xy)\,-\,2y\,$$<br>
$$\hspace{2 em}\Longrightarrow h\,'(y)\,=\,-2y\,$$<br>
This last expression can be integrated to yield $\,h\,=\,-y^2\,+\,c_1\,$<br><br>
Substituting this $\,h\,$ into the "unfinished g" line, results in<br><br>
$$g(x,y)\,=\,2x\,+\,\textrm{exp}(xy)\,-\,y^2\,+\,c_1\,$$<br>
Pulling the $\,g(x,y) = c\,$ primordial equation out of the equation locker, gives a finish to the search with the solution<br><br>
$$\,2x\,+\,\textrm{exp}(xy)\,-\,y^2\,=\,c_2\,$$<br><br>
As in previous problems, the solution is implicit. The matplotlib treatment can be realized as shown below.



     
</font>                 
</div>

<font color='blue'> The M blob was identified as $\,y^2\,$, and already has the $\,dt\,$ indigenous differential associated with it. The first task is to integrate M with respect to t:<br><br>
$$\int\,\frac{\partial g}{\partial t}\,dt\,=\,\int\,y^2\,dt\,$$<br>
$$\hspace{2 em}\Longrightarrow g(y,t)\,=\,y^2t\,+\,h(y)\,\hspace{4 em}$$<br>
Then to differentiate the last line with respect to $\,y\,$<br><br>
$$\,2yt\,+\,\frac{dh}{dy}\,$$<br>
Since it has the $\,dy\,$ differential associated with it, it is bound to be equal to N. Therefore,<br><br>
$$2yt\,+\,\frac{dh}{dy}\,=\,2yt\,+\,1\,$$<br>
The last equation can be thought of as the "unfinished $\,g(x,\,y)\,$".<br><br>
By inspection, $\,\large\frac{dh}{dy}\normalsize\,=\,1\,$, therefore $\,h(y)\,=\,y\,+\,c_1\,$ and the "unfinished $\,g(x,\,y)\,$" becomes<br><br>
$$\,g(t,\,y)\,=\,y^2t\,+\,y\,+\,c_1\,$$<br><br>
Pulling out the old primeaval expression of $\,g(x,\,y)\,$ results in a solution of<br><br>
$$\,y^2t\,+\,y\,=c_2\,$$<br>
As it happens, the last line can be solved explicitly for $\,y\,$ using the quadratic equation.






     
</font>                 
</div>

```python
from sympy import *
y, t, c2 = symbols("y t c2")

solve(y**2*t + y - c2, y)

```

<font color='blue'> Multiplying the stated equation by the stated factor gives the expression:<br><br>
$$\,\frac{-1}{x^2}\,(y\,dx\,-\,x\,dy)\,=\,0\,$$<br>
$$\hspace{4 em}\Longrightarrow \frac{-y}{x^2}\,dx\,+\,\frac{1}{x}\,dy\,=\,0\,\hspace{6 em}$$<br>
The form of the blobs in the last line is the right form for applying the test for equation exactness, with $\,M\,=\,\large\frac{-y}{x^2}\,$ and $\small\,N\,=\,\large\frac{1}{x}\,\small$<br><br>
Taking the partial of M with respect to $\,y\,$ gives $\,\large\frac{-1}{x^2}\,$. Taking the partial of N with respect to $\small\,x\,$ gives $\large\,\frac{-1}{x^2}\,$.<br><br>
The partials being equal means that the equation is exact, and that, therefore, the factor $\,\large\frac{-1}{x^2}\,$ is a integrating factor for the starting equation.
     
     
     

</font>                 
</div>

<font color='blue'>5.18. Solve $\,(y^2\,-\,y)\,dx\,+\,x\,dy\,=0\,$.

<div style="background: #E2FFCC;

<font color='blue'> The equation as given is not exact. The text advises that a certain grouping of the elements will be helpful.<br><br>
$$\,-\,(y\,dx\,-\,x\,dy)\,+\,y^2\,dx\,=\,0\,$$<br>
The purpose of this regrouping is to get the $\,(y\,dx\,-\,x\,dy)\,$ factor isolated. This factor is interesting because four different integrating factors are designed to fit with it. Guidance points to the integrating factor $\,\large\frac{1}{y^{\,2}}\,$ as the one which can unlock the exactness potential of the problem expression.<br><br>
$$\,\Longrightarrow\frac{1}{y^2}\,(-\,(y\,dx\,-\,x\,dy)\,+\,y^2\,dx)\,=\,0\,$$<br>
$$\,\Longrightarrow\frac{-\,y\,dx\,+\,x\,dy\,}{y{\,^2}}\,+\,1\,dx\,=\,0\,\hspace{2 em}$$<br>
The expression on the last line needs to be rearranged into differential form in order to judge its characteristics.<br><br>
$$\,(1\,-\,\frac{1}{y})\,dx\,+\,\frac{x}{y^{\,2}}\,dy\,=\,0\,$$<br>
$$\hspace{4 em}\large M\hspace{4 em}N\small\,\hspace{6 em}$$<br><br>
Taking the partial derivative of $\,M\,$ with respect to $\,y\,$ and the partial derivative of $\,N\,$ with respect to $\,x\,$, produces the forms<br><br>
$$\hspace{2 em}\,\frac{1}{y^{\,2}}\:\:\:\:\:=\:\:\:\:\:\frac{1}{y^{\,2}}\hspace{4 em}\,$$<br>




</font>                 
</div>

<div style="background: #E2FFCC;
            font-family: Bookerly;

            border: 30px solid #E2FFCC; 
            margin-left: 1px;
            margin-right: 1px;"

<font color='blue'>So the equation is now exact. To go on to the actual solution, the first task is to integrate the blob $\,M\,$ with respect to its indigenous differential, $\,dx\,$<br><br>
$$\,\int\,(1 \,-\,\frac{1}{y})\,dx\,$$<br><br>
$$\,\hspace{2 em}\Longrightarrow x\,-\,\frac{x}{y}\,+\,h(y)\hspace{4 em}\,$$<br>
Then to differentiate the resulting expression with respect to y:<br><br>
$$\,\hspace{2 em}\Longrightarrow\:\:\: \,\frac{x}{y^{\,2}}\,+\,h\,'(y)\hspace{4 em}\,$$<br>
At this point the developed expression needs to match N, so to set that up<br><br>
$$\,\hspace{2 em}\Longrightarrow\:\:\:\frac{x}{y^{\,2}}\,=\,\frac{x}{y^{\,2}}\,+\,h\,'(y)\hspace{4 em}\,$$<br>
This implies that $\,h\,'(y)\,$ is equal to zero, thus $\,h(y)\,$ merely a constant, and that the final solution is the one found three lines up, modified with the new intelligence about $\,h(y)\,$<br><br>
$$\,\hspace{2 em}\Longrightarrow x\,-\,\frac{x}{y}\,+\,h(y)\,=\,x\,-\,\frac{x}{y}\,+\,c$$<br>
The plot below captures the function complemented by a family of constant values.



</font>                 
</div>

<font color='blue'> The equation as given is not exact. The text advises that a certain grouping of the elements will be helpful.<br><br>
$$\,(y\,dx\,+\,x\,dy)\,+\,(-xy^{\,2}\,dx\,+\,x^2y^{\,2}\,dy)\,=\,0\,$$<br>
The purpose of this regrouping is to get the $\,(y\,dx\,+\,x\,dy)\,$ factor isolated. This factor is interesting because four different integrating factors are designed to fit with it. Guidance points to the integrating factor $\,\large\frac{1}{(xy)^{\,2}}\,$ as the one which can unlock the exactness potential of the problem expression.<br><br>
$$\,\Longrightarrow\frac{1}{(xy)^2}\,((y\,dx\,+\,x\,dy)\,+\,(-xy^{\,2}\,dx\,+\,x^2y^{\,2}\,dy))\,=\,0\,$$<br>
$$\,\Longrightarrow\frac{y\,dx\,+\,x\,dy}{(xy)^2}\,+\,\frac{-xy^{\,2}\,dx\,+\,x^2y^{\,2}\,dy}{(xy)^2}\,=\,0\,$$<br>
The expression on the last line can be reduced in complexity, into what may be called the "bailout line".<br><br>
$$\,\Longrightarrow\frac{y\,dx\,+\,x\,dy}{(xy)^2}\,+\,\frac{-1\,dx}{x}\,+\,1\,dy\,=\,0\,$$<br>
The expression on the last line needs to be rearranged into a suitable differential form in order to judge its characteristics.<br><br>
$$\,\Longrightarrow\left(\frac{y}{(xy)^{2}}\,-\,\frac{1}{x}\right)\,dx\,+\,\left(\frac{x}{(xy)^2}\,-\,1\right)\,dy\,=\,0\,$$

$$\hspace{4 em}\large M\hspace{8 em}N\small\,\hspace{6 em}$$<br>

Sympy can help out with the exactness test.



</font>

```python
from sympy import symbols, diff
x, y = symbols('x y', real=True)
M = (y/(x*y)**2) - 1/x
ml = diff(M, y)
ml
```
