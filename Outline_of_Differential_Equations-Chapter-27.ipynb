<font color='blue'>27.2 Find a recurrence formula for the power series solution around $\:x\,=\,0\,$ for

$$y\,''\,-\,xy\,'\,+\,2y\,=\,0\,$$

</font>

```python
import sympy as sp
x = sp.Symbol('x')
y = sp.Function('y')(x)
yp = sp.Derivative(y, x)
ydp= sp.Derivative(y, x, x)
equation = sp.Eq(3*x + yp + 2*y, 0)
display(equation)

```

```python
solution = sp.dsolve(equation, hint='1st_power_series')
display(solution)

```

<font color='blue'>If this differential equation is entered in Symbolab, the solution which is returned is:

$$y \,=\,c_1\,\textrm{cos}\,(t)\,+\,c_2\,\textrm{sin}\,(t)$$

The Maclaurin series for these two trig functions are equal to:

$$\textrm{cos}\,(x)\,=\,1\,-\,\frac{x^2}{2!}\,+\,\frac{x^4}{4!}\,-\,\cdot\cdot\cdot\,=\,\sum_{n=0}^\infty\frac{(-1)^n\,x^{2n}}{(2n)!}$$

$$\textrm{sin}\,(x)\,=\,1\,-\,\frac{x^3}{3!}\,+\,\frac{x^5}{5!}\,-\,\cdot\cdot\cdot\,=\,\sum_{n=0}^\infty\frac{(-1)^n\,x^{2n\,+\,1}}{(2n\,+\,1)!}$$

No method for derivation of the recurrence formula based on the above information can be discerned. The derivation process, as developed by the text, is fairly involved.

!| solve[[y'' + x y' + (2 x -1) y = 0],{y(-1) = 2, y'(-1) = -2}] |!

and the result looks like:

$$\begin{align}
  &{y(x)\,=\,e^{-1/2(x-4)x\ -\ 29/2} \,\left( -43 e^{9/2}\,\sqrt{2\pi}\,(x^2\,-\,8x\,+\,15)\,\mathsf{erfi}\,\left(\frac{x\,-\,4}{\sqrt{2}}\right)\,-\,43e^{9/2}\sqrt{2\,\pi}\,\mathsf{erfi}\,\left(\frac{5}{\sqrt{2}}\right)\,+ \right.} \\

One thing that would seem sensible would be to get a numerical evaluation of the function at the first initial value coordinate. Unfortunately, this cannot be done in Wolfram Alpha because the number of characters to enter the expression would exceed the number allowed. What can be done is to cut the primary expression into three parts, evaluate each, and then join the numerical results.

① !| [e^(-(1/2) ((-1) - 4) (-1) - (29/2)) (-43 e^(9/2) sqrt(2 pi) ((-1)^2 - 8 (-1) + 15) erfi(((-1) - 4)/sqrt(2)))] |!

<font color='blue'>This problem is a step full into Airyland. The entry line in Wolfram Alpha is:

!| solve[[y'' - 2 x y = 0],{y(2) = 1, y'(2) = 0}] |!

and the return is:

$$y(x) \,=\,\small\mathsf{\frac{Ai'(2\,\sqrt[3]{2})\,Bi\,(\sqrt[3]{2}\,x)\,-\,Bi'\,(2\,\sqrt[3]{2})\,Ai\,(\sqrt[3]{2}\,x)}{Ai'\,(2\,\sqrt[3]{2})\,Bi\,(2\,\sqrt[3]{2}\,)\,-\,Ai\,(2\sqrt[3]{2}\,)\,Bi'\,(\,2\sqrt[3]{2}\,)}}$$

On substituting the value $x\,=\,2$, W|A compliantly produces the value 1, in agreement with the first initial condition.

Wolfram Alpha is also ready to provide a derivative for the above equation (if care is taken in putting in the entry):

!|  D[(Ai' (2 2^(1/3)) Bi (2^(1/3) x) - Bi' (2 2^(1/3)) Ai (2^(1/3) x))/(Ai' (2 2^(1/3)) Bi (2 2^(1/3)) - Ai (2 2^(1/3)) Bi' (2 2^(1/3)))] |!

for which the answer is:

$$y\,'(x) \,=\,\sqrt[3]{2}\,\pi \left( \mathsf{Bi'} \left( 2\sqrt[3]{2} \right) \mathsf{Ai'}\left(\sqrt[3]{2}x\right)\,-\, \mathsf{Ai'} \left(2 \sqrt[3]{2} \right) \mathsf{Bi'}\left(\sqrt[3]{2} x\right)\right)$$

(The above is listed as an alternate form of the answer, and is the shortest version of three.)

It is possible to check the second initial condition within W|A inself. The front and back ends of the input above are changed slightly, so that the entire line reads:
!| ReplaceAll[ D[(Ai' (2 2^(1/3)) Bi (2^(1/3) x) - Bi' (2 2^(1/3)) Ai (2^(1/3) x))/(Ai' (2 2^(1/3)) Bi (2 2^(1/3)) - Ai (2 2^(1/3)) Bi' (2 2^(1/3)))] , {x->2}] |!

And $0$ is returned as the answer, verifying the second initial condition.

```python
import numpy as np
import sympy as sp
from sympy import airyai
from sympy import airybi
import matplotlib.pyplot as plt

%config InlineBackend.figure_formats = ['svg']

x = sp.symbols('x')

x1 = (sp.diff(sp.airyai(x))*2*2**(1/3)*airybi(x)*2**(1/3)*x - \
          sp.diff(sp.airybi(x))*2*2**(1/3)*airyai(x)*2**(1/3)*x)/\
          (sp.diff(sp.airyai(x))*2*2**(1/3)*airybi(x)*2*2**(1/3) - \
          sp.airyai(x)*2*2**(1/3)*sp.diff(airybi(x))*2*2**(1/3))

x2 = 2**(1/3)*sp.pi*(sp.diff(sp.airybi(x))*2*2**(1/3)*sp.diff(sp.airyai(x))*2**(1/3)*x -\
          sp.diff(sp.airyai(x))*2*2**(1/3)*sp.diff(sp.airybi(x))*2**(1/3)*x)
          
            
xt = np.arange(0, 4, 0.01)
yt1 = np.zeros(len(xt))
yt2 = np.zeros(len(xt))

for n in range(0,len(xt)):
    yt1[n] = sp.N(x1.subs(x, xt[n]))
    yt2[n] = sp.N(x2.subs(x, xt[n]))
    
plt.grid(True, linestyle='dotted', color='#999999', linewidth=0.9)
plt.xlabel("x")
plt.ylabel("y(x)")
plt.title("Outline of Differential Equations Prob 27.24")
plt.rcParams['figure.figsize'] = [5,5]
plt.rcParams['font.serif'] = ['Bookerly']
plt.rcParams['mathtext.fontset'] = 'dejavuserif'

ax = plt.gca()
ax.axhline(y=0, color='#CCCCCC', linewidth=0.7)
ax.axvline(x=0, color='#CCCCCC', linewidth=0.7)
ratio = 0.98
xleft, xright = ax.get_xlim()
ybottom, ytop = ax.get_ylim()
ax.set_aspect(abs((xright-xleft)/(ybottom-ytop))*ratio)

plt.text(0.5, 1.3, " y(x) ", size=10,\
         bbox=dict(boxstyle="square", ec=('#6666BB'),fc=(1., 1., 1),))
plt.text(0.5, -0.75, " y'(x) ", size=10,\
         bbox=dict(boxstyle="square", ec=('#CC6666'),fc=(1., 1., 1),))
xpts = np.array([2, 2])
ypts = np.array([1, 0])
plt.plot(xpts, ypts, markersize=7, color='k', marker='s', \
         mfc='none', linestyle = 'none', markeredgewidth=0.5)
plt.ylim(-2, 4)
plt.xlim(-2, 5)
plt.plot(xt, yt2, linewidth = 0.9, color = '#CC6666')
plt.plot(xt, yt1, linewidth = 0.9, color = '#6666BB')
plt.show()

```
