<font color='blue'>Test for exactness: If M(x,y) and N(x,y) are continuous functions and have continuous first partial derivatives on some rectangle of the xy-plane, then (a) $or$ (b) is exact if and only if<br>

$$\frac{\partial M(x,y)}{\partial y}\,=\,\frac{\partial N(x,y)}{\partial x}$$<br>

Though in the symbolic equation the M blob is associated with $\,dy\,$, it is identified in the problem equation by an association with $\,dx\,$.<br>

So in equation (a), the partial derivative of the first blob with respect to y is: $\,3x^2\,$<br>

And also in equation (a), the partial derivative of the second blob with respect to x is: $3x^2$<br>

Note that the indigenous differentials in both blobs are ignored. It is assumed for the present that their only role is in identifying whether an x-blob or a y-blob is being examined.

The decisive test partials being found equal, it is concluded that equation (a) is exact.

How about for equation (b)? Here the partial derivative of the first blob with respect to y is: $\,x\,$

And also for equation (b), the partial derivative of the second blob with respect to x is:
$\,0\,$

The decisive test partials being found to be unequal, it is concluded that equation (b) is not exact.

<font color='blue'>$$\,y\,'\,=\,y/x\,\,$$

The text answer on this one is interesting. Pointing out that only equations in differential form can be judged for exactness, the text examines two possible differential forms:<br>

$$\,\frac{y}{x}\,+\,(-1)dy\,=\,0\,$$<br>

with the test equation equaling:<br>
$$\,\frac{\partial{M}}{\partial{y}}\,=\,
\frac{1}{x}\,\ne\,0\,=\,\frac{\partial{N}}{\partial{x}}$$

for which case, the decisive test failing, the equation must be judged not exact. But the second differential form:<br>

$$\,\frac{1}{x}\,dx\,+\,\frac{-1}{y}\,dy\,=\,0$$<br>

leads to the decisive test<br>
$$\,\frac{\partial{M}}{\partial{y}}\,=\,0\,=\,\frac{\partial{N}}{\partial{x}}$$<br>

for which case, the decisive test succeeding, the same equation, in a different differential form, must be judged exact.
