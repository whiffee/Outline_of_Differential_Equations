<font color='blue'>27.2 Find a recurrence formula for the power series solution around $\:x\,=\,0\,$ for

$$y\,''\,-\,xy\,'\,+\,2y\,=\,0\,$$

```python
import sympy as sp
x = sp.Symbol('x')
y = sp.Function('y')(x)
yp = sp.Derivative(y, x)
ydp= sp.Derivative(y, x, x)
equation = sp.Eq(3*x + yp + 2*y, 0)
display(equation)

```

```python
solution = sp.dsolve(equation, hint='1st_power_series')
display(solution)

```

```python
from sympy import *

x = symbols('x')
n = Symbol('n', positive=True, integer=True)
r = limit((-x**(2*n-1))/factorial(2*n - 1), x, oo)
if (r==0):
    print('nope')
else:
    print('converges to  ' + str(r))
print('limit of the given function  ' + str(r))         
         
    
```

<font color='blue'>If this differential equation is entered in Symbolab, the solution which is returned is:

$$y \,=\,c_1\,\textrm{cos}\,(t)\,+\,c_2\,\textrm{sin}\,(t)$$

The Maclaurin series for these two trig functions are equal to:

$$\textrm{cos}\,(x)\,=\,1\,-\,\frac{x^2}{2!}\,+\,\frac{x^4}{4!}\,-\,\cdot\cdot\cdot\,=\,\sum_{n=0}^\infty\frac{(-1)^n\,x^{2n}}{(2n)!}$$

$$\textrm{sin}\,(x)\,=\,1\,-\,\frac{x^3}{3!}\,+\,\frac{x^5}{5!}\,-\,\cdot\cdot\cdot\,=\,\sum_{n=0}^\infty\frac{(-1)^n\,x^{2n\,+\,1}}{(2n\,+\,1)!}$$

No method for derivation of the recurrence formula based on the above information can be discerned. The derivation process, as developed by the text, is fairly involved.

!| solve[[y'' + x y' + (2 x -1) y = 0],{y(-1) = 2, y'(-1) = -2}] |!

and the result looks like:

$$\begin{align}

One thing that would seem sensible would be to get a numerical evaluation of the function at the first initial value coordinate. Unfortunately, this cannot be done in Wolfram Alpha because the number of characters to enter the expression would exceed the number allowed. What can be done is to cut the primary expression into three parts, evaluate each, and then join the numerical results.
