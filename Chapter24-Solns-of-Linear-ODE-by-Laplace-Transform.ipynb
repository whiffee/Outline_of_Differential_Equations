<font color='blue'>Note that in the first input line above, all $x$ factors were converted to $t$ factors to make the process work. With the transform available, Wolfram Alpha can take it from here.

!| solve[s^3 y + s y = 1/(s -1)], y |!

gives the solution to $Y(s)$ in s-space, which is

$\,y\,=\,\large \frac{1}{s\,(s^3\,-\,s^2\,+\,s\,-\,1)}$



Then

!| inverselaplacetransform[1/(s (s^3 - s^2 + s - 1)) |!

takes the solution back to t-space, resulting in

$\large \frac{1}{2} \small (e^t\,-\,\textrm{sin}\,(t)\,+\,\textrm{cos}\,(t)\,-2\,)$

or, un-substituting,

$\large \frac{1}{2} \small (e^x\,-\,\textrm{sin}\,(x)\,+\,\textrm{cos}\,(x)\,-2\,)$
